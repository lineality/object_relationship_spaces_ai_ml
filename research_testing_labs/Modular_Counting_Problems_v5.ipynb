```python
colours = ["Red",
"Blue",
"Green",
"Yellow",
"Orange",
"Purple",
"Pink",
"Brown",
"Black",
"White",
"Gray",
"Teal",
"Cyan",
"Maroon",
"Indigo",
"Lavender",
"Gold",]
```

```python
#####################
# Systematic Counter
#####################

input = colours

def get_starting_letter_frequency(strings_list):
    frequency_dict = {}

    for string in strings_list:
        if string:
            first_letter = string[0].lower()  # Convert to lowercase to handle case-insensitivity

            # print(first_letter)

            if first_letter.isalpha():
                frequency_dict[first_letter] = frequency_dict.get(first_letter, 0) + 1

    return frequency_dict

# Example usage
frequency_result = get_starting_letter_frequency(input)
print(frequency_result)

def get_shared_values(input_dict):
    shared_values_dict = {}

    for key, value in input_dict.items():
        if value in shared_values_dict:
            shared_values_dict[value].append(key)
        else:
            shared_values_dict[value] = [key]

    shared_values_dict = dict(sorted(shared_values_dict.items(), key=lambda item: item[0], reverse=True))
    return shared_values_dict

frequency_result
value_counts = get_shared_values(frequency_result)

for value, keys in value_counts.items():
    print(f"Keys with value {value}: {keys}")

def top_key_count_values(dictionary):
    if dictionary:
        first_key = next(iter(dictionary))
        return len(dictionary[first_key])
    else:
        return 0

first_key_count = top_key_count_values(value_counts)
print(f"\nThe number of first key values: {first_key_count}")

def get_second_key(dictionary):
    keys = list(dictionary.keys())
    if len(keys) >= 2:
        return keys[1]
    else:
        return None

second_key = get_second_key(value_counts)
print(f"The second key in the dictionary is: {second_key}")

X = second_key + first_key_count

print(f"Sum of second key and first key value-count: {X}")

Y = len(input)

solution = Y - X


print(f"\nY -> {Y}")
print(f"X -> {X}")
print(f"solution -> {solution}")
```
